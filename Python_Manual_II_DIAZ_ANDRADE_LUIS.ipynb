##**I. FAIR Principles**
---



<img src="https://www.escueladerevisores.com/wp-content/uploads/2020/12/Imagen1.jpg" width="500px">

*(FAIR Principles. Government of Spain. Ministry of Economic Affairs and Digital Transformation, 2023)*



(Wilkinson et al., 2016). The FAIR Principles are a set of guiding principles for making research data easy to find, accessible, interoperable, and reusable. These principles provide guidance for scientific data management and directly address data producers and data publishers to promote maximum use of research data.

{Alcal√°, M., & Anglada, L. (2019)}. The most relevant elements that are integrated by the FAIR principles are:

1. Documents and Metadata: They allow data to be findable, reproducible, reusable and interoperable. The documentation takes into account the description of the collection process, provenance and the data itself (metadata).
2. Data formats: These define the extent of interoperability and reusability of the data. If possible, opt for an open format, contain embedded metadata and be sustainable.
3. Data access: In the first instance, do not publish confidential data. Plan, at the beginning of the project, what data can be shared and how to share it. If data cannot be shared, at least provide access to data metadata is expected to be provided.
4. Persistent identifiers: Allows data to be accurately findable, accessible and citable. In the case of data, there are Digital Object Identifiers (DOIs).
Open data licensing: Allows data to be distributed freely or with as few restrictions as possible. It is recommended to opt for Creative Commons CC0 or CC0-BY 4.0.

(Wilkinson, M. D., et al. 2019) The "FAIR" principle is a set of principles to ensure that data is easily found, accessible, interoperable and reusable. Some examples of how these principles can be applied:

1. Findable:

 Provide a unique and permanent identifier for data (e.g., a DOI or digital object identifier).
Use clear and accurate descriptive metadata to enable the data to be easily searched and found in catalogs or search engines.
2. Accessible:

 Provide access to data free of charge or at a reasonable cost.
Ensure that data is accessible to people with disabilities through the use of accessible technologies.
3. Interoperable:

 Use standardized and widely accepted data formats to allow data to be shared and used by different systems.
Include sufficient information in metadata to allow data to be interpreted correctly.
4. Reusable:

 Publish the data under a clear license that allows the data to be reused and redistributed.
Provide the data in a format that is easily processed and manipulated.
By following these "FAIR" principles, the data become more useful and valuable to the scientific community and to other users who wish to use them.
 

##**II. Documentation & Metadata**

**II.1 Metadata**

Metadata is structured information that describes, explains, locates, or makes it easier to retrieve, use, or manage an information source. easier to retrieve, use or manage a source of information. Metadata is often referred to as data that is used to describe other data.

Examples of metadata

1. Photographs and images. When taking a picture, many digital cameras include a series of metadata useful for further processing: shutter speed, aperture opening, focal length, chosen mode (automatic or manual), ISO speed, coordinates of the place where the shot was taken (if the camera is equipped with GPS), the camera model used, and so on. Subsequently, the photographer can add others such as the identity of the author, description of the scene or the corresponding keywords, which will make it easier to search in a database.

2. Written materials. In the form of metadata, the length (number of pages, words or characters), the content (a summary, an index of chapters, keywords, topics covered), the author, the date of creation, the type of content (whether it is a literary work, and if so, the genre to which it belongs, or a technical manual, or a legal text), and so on, can be displayed.

3. Videos. Some useful metadata in a recording, which can be added to those used in still photography, refer to the format in which they are recorded, their duration, their bitrate (bit rate: amount of information per unit of time, which will give you an idea of both their quality and file size), the date of recording and editing, the software used, the codec needed to play them, and so on. Much of this metadata is also used for sound-only recordings.

##**III. Data Storage & Organization**

**III.1 Data Storage**

Data storage refers to the process of saving information on a storage device for later use. There are different types of storage devices, each with its own advantages and disadvantages.

1. Hard disk: is a non-volatile storage device that uses magnetic disks to store data. It is widely used in desktop and laptop computers, and offers a large storage capacity and reasonably fast speed.

2. Flash memory: is a type of non-volatile storage that uses memory chips to store data. It is widely used in portable devices such as smartphones, digital cameras and USB flash drives due to its compact size, low power consumption and high speed.

3. Magnetic tape drives: are data storage devices in which data is stored on magnetic tapes. They are widely used in the audio and video recording industry, as well as in large-scale data backup.

4. Solid State Drives (SSD): are non-volatile storage devices that use flash memory chips to store data. They offer much faster speed than hard disks and are widely used in servers, workstations and laptops.

5. Cloud storage: is a form of online data storage that allows users to store and access their data over the Internet. Examples of cloud storage services include Dropbox, Google Drive, OneDrive and iCloud.

```python
import pandas as pd
import numpy as np

df_Sc = pd.DataFrame({"Alumno1": np.array([100,70,80, np.nan, 50, 100]), "Alumno2": np.array([60,np.nan,80,90,100, 60]), "Alumno3": np.array([100, 100, 100, 80, np.nan, 90]), 
                          "Alumno4": np.array([70, 50, 90, 50, 100, 70])})

#Function np.nan is an empty celtic 
#(In this example, np.nan is not aproviedd)
df_Sc
```

```python
'''Then, here we have 2 options: delete the rows that contain NaN values or replace these NaN values by something that has some kind of sense in the dataframe.

The first option is the easiest, but it implies losing the whole row where in other columns there may be values:'''

df_manual1 = df_Sc.copy()
df_manual1.dropna(how="any")

#The how parameter can be "any" or "all". If it is the former, the entire row or column is deleted if it contains at least
# one NaN value. If "all", the row or column is deleted only if ALL elements are NaN. 
```

```python
'''
Now let's look at the "difficult" option. To replace these values there are many options, some more sophisticated than others:

Replace by some specific value (a scalar).
Replace by the average/median of the column
Replace with some built-in method offered by Pandas
Replace using some algorithm; for example, some algorithms that do it automatically.


'''

df_manual2 = df_Sc.copy()
df_manual2.fillna(70)
```

```python
# We replace with the average of each column
df_manual2 = df_Sc.copy()
df_manual2.fillna(df_manual2.mean())
```

###**XI. Concatenate dataframes**
---

The function to concatenate dataframes. Taking into account that we have a database of red and white wines, it is of interest to merge them and add a categorical variable hue, which can be used as a basis for the concatenation of dataframes.

```python
import pandas as pd
import numpy as np

df_Sc = pd.DataFrame({"Alumno1": np.array([100,70,80, 90, 50, 100]), "Alumno2": np.array([60,np.nan,80,90,100, 60]), "Alumno3": np.array([100, 100, 100, 80, np.nan, 90]), 
                          "Alumno4": np.array([70, 50, 90, 50, 100, 70])})

df_Sc['hue']='GROUPA'



df_Scb = pd.DataFrame({"Alumno1": np.array([90,70,80, 90, 59, 86]), "Alumno2": np.array([89,56,80,90,100, 60]), "Alumno3": np.array([100, 89, 100, 90, 73, 90]), 
                          "Alumno4": np.array([74, 50, 87, 80, 40, 70])})

df_Scb['hue']='GROUPB'

df_im = pd.concat([df_Sc, df_Scb], ignore_index=True)
df_im
```

```python
#We count and display the amount of data per group.

df_im.value_counts("hue")
```

###**XI. Dinamic Tables**
---



It is of interest to have 2 categorical variables, hue and quality, to make a pivot table as in Excel.

Pivot tables (or pivot tables) are a common tool in spreadsheets such as Microsoft Excel or Google Sheets that allow you to summarize large amounts of data in a table that is easier to analyze.

For this, we need to define:

* The rows or index
*The columns
*The values that can be averages, maximum values, minimum values, sums, etc.

```python
#It may be convenient to define a new column that is an average of units 4, 5 and 6.

df_im["PROMEDIO"] = (df_im["Alumno1"] + df_im["Alumno2"]+ df_im["Alumno3"]+ df_im["Alumno4"])/4
df_im["CALIF"] = (df_im["Alumno1"] + df_im["Alumno2"]+ df_im["Alumno3"]+ df_im["Alumno4"])/5


df_im
```

```python
pivot_table = df_im.pivot_table(index='PROMEDIO', columns='hue', values='CALIF')
pivot_table
```

```python
pivot_table["GROUPA%"] = (pivot_table["GROUPA"]/6)
pivot_table["GROUPB%"] = (pivot_table["GROUPA"]/6)
pivot_table
```
