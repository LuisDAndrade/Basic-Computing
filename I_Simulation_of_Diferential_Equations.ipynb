```python
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
```

```python
tst = 0
tend = 3
inc = 0.1
g = 9.81

x0 = 0
t = np.arange(tst,tend+inc,inc)
```

```python
def mydiff(x,t,g):
  dxdt = -g*t + 10
  return dxdt
```

```python
x = odeint(mydiff,x0,t,args=(g,))
print(x)


hmax = [100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81),100/(2*9.81)]

tf = [2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81),2*(10/9.81)]
tm = [10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81,10/9.81]
```

```python
figure = plt.figure(figsize=(10,10))
ax = figure.add_subplot(111)
ax.plot(t,x,color='purple')
ax.plot(tm, x, color='yellow', label=tm )
ax.plot(tf, x, color='pink')
ax.plot(hmax, color='green')
ax.set_title('Plotting Differential Equiation',size=18,color='y')
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.axis([-0.25,2.5,-0.25,6])
plt.grid(color='m', linestyle='dotted', linewidth=1)
plt.legend
plt.show
```
