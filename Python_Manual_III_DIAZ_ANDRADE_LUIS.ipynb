##**vi. Boxplots**
---

**Box-and-whisker plots (also known as boxplots) are a type of graph used to represent the distribution of a set of numerical data.**

**The plot consists of a rectangle (the "box") extending from the first quartile (Q1) to the third quartile (Q3) of the data set. Within the box, a vertical line is drawn representing the median (the value that divides the data set into two equal parts).**

**From the box, diagonal lines extending outward, called "whiskers," are drawn extending to the minimum and maximum values of the data set that are not outliers. Outliers are plotted as individual points beyond the whiskers.**

**In summary, the box-and-whisker plot provides a visual way to view the variability, shape, and outliers of a data set. It is useful for identifying outliers and comparing distributions between different data sets.**

**We can make this type of plots for each variable, this type of plot allows us, among other things, to easily visualize the outliers**


```python
#Graph boxplot in Seaborn:

sns.boxplot(y='PROMEDIO',x="Eq", data = df,orient="v")

#title, axis titles, and tick size:
plt.title("TRAILS AVARAGE", fontsize=18)
hist.tick_params(labelsize=14)
```

##**ix. Pearson's Correlation Coefficient**
---

We can calculate the correlation between two variables with the Pearson correlation coefficient and the p-value.

**The Pearson correlation coefficient ( œÅ ) allows us to easily observe the linear dependencies between two datasets.**

**A value of +1 means total positive linear correlation, 0 means no linear correlation, and -1, total negative linear correlation.**
**Positive correlations imply that as x increases, so does y. Negative correlations imply that as x increases, y increases. Negative correlations imply that as x increases y decreases.**

**The p-value indicates approximately the probability that the uncorrelated data sets have a Pearson correlation coefficient at least as extreme as that determined for those data sets. If this probability is less than 5% (P<0.05), it means that the correlation coefficient is statistically significant.**

**Care must be taken because correlation does not imply causation. When two variables are correlated, there may or may not be a causal connection, and this connection may also be indirect.**

```python
#I import the PearsonR function from the "stats" module SciPy library
from scipy.stats import pearsonr
def get_corr(col1, col2, temp_df):
    pearson_corr, p_value = pearsonr(temp_df[col1], temp_df[col2])
    print("The correlation between {} and {} is {:.2f}".format(col1, col2, pearson_corr))
    print("The P-value of this correlation is {:.1e}".format(p_value))
get_corr('AVANCE1', 'CALIFU4', df)
```

##**x. Heatmap**
---

**We can create a DataFrame with the Pearson correlation coefficient for each pair of columns (using the df.corr() function) and then use Matplotlib and Seaborn to create a heatmap of the coefficients.**

```python
# We calculate the correlation with Pearson's coefficient using Pandas:
correlation = df2.corr(method='pearson')
# Create a new figure with Matplotlib and set the size:
fig=plt.gcf()
fig.set_size_inches(9,9)
# correlation heatmap using Seaborn:
sns.heatmap(correlation, annot=True,square=True)
```

**When several variables are compared at the same time it is difficult to understand how the data behave. So we could add to the heatmap the coefficients of each comparison and plot only the lower triangular correlation matrix (since it is a symmetric plot, and the information in the upper triangle is redundant).**

```python
import matplotlib.pyplot as plt
import numpy as np
correlation = df2.corr()
# np.triu es una funci√≥n de Numpy que agarra un array en forma de matriz cuadrada y me da el tri√°ngulo superior de esa matriz cuadrada
upper_triangle = np.triu(correlation)
fig=plt.gcf()
fig.set_size_inches(10,10)
sns.heatmap(data=correlation,mask=upper_triangle,square=True,annot=True)
```
