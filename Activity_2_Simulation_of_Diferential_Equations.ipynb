$m(\ddot{x}) +b(\dot{x}) +k(x) = 0 $ 

$m(\ddot{x}) = - b(\dot{x}) - k(x)$

$(\ddot{x}) = -(\frac{b}{m}) (\dot{x}) - (\frac{k}{m})x$

```python
!pip install scipy
```

```python
#We import necessary libraries 

import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
```

```python
k = 81.75
m = 2
b = 0.5
```

```python
def mydiff1(x,t,m,k,b):

  dx1dt = x[1]
  dx2dt = ((-b/m)*x[1]) - ((k/m)*x[0])
  dxdt = [dx1dt,dx2dt] 
  return dxdt
```

```python
tst = -5
tend = 25
inc = 0.1
t = np.arange(tst,tend+inc,inc)
xinit = [1,0]
x = odeint(mydiff1,xinit,t,args=(m,k,b,))
```

```python
x1 = x[:,0]
x2 = x[:,1]

figure = plt.figure(figsize=(10,7))
plt.plot(t,x1,label=r'$x$')
plt.plot(t,x2,label=r'$\dot{x}$',color='violet')
plt.grid(color='m', linestyle='dotted', linewidth=1)
plt.legend()
plt.show()
```

$1.$ $m(\ddot{x_1}) + k_1({x_1-x_2}) +k_2(x_1) = 0 $ 

$2.$ $k_1(x_1-x2) = b_1 \dot{x_2}$

---
$1.$ $m(\ddot{x_1}) + k_1({x_1-x_2}) +k_2(x_1) = 0 $ 

$ $ $ $ $1.1$ $m(\ddot{x_1}) = - k_1({x_1-x_2}) - k_2(x_1)  $ 

$---> $ $ $ $1.2$ $\ddot{x_1} = - \frac{k_1}{m}({x_1-x_2}) - \frac{k_2}{m}(x_1)  $ 

---


$2.$ $k_1(x_1-x2) = b_1 \dot{x_2}$

$---> $ $ $ $ 2.1 $ $\dot{x_2} = \frac{k_1}{b_1} (x_1 - x_2)$

```python
k1 = 50
k2 = 31.75
m1 = 0.2
b1 = 0.4
```

```python
def mydiff2(x,t,m1,k1,k2,b1):

  dx1dt = x[1]
  dx2dt = ((-k1/m1)*(x[0]-x[1]) - ((k2/m)*x[0]))
  dx3dt = ((k1/b1) * (x[0]-x[1]))
  dxdt = [dx1dt,dx2dt,dx3dt] 
  return dxdt
```

```python
tst = -1
tend = 10
inc = 0.1
t = np.arange(tst,tend+inc,inc)
xinit = [1,0,0]
x = odeint(mydiff2,xinit,t,args=(-b1,m1,k1,k2,))
```

```python
x1 = x[:,0]
x2 = x[:,1]


figure = plt.figure(figsize=(10,7))
plt.plot(t,x1,label=r'$x$')
plt.plot(t,x2,label=r'$\dot{x}$',color='green')
plt.grid(color='m', linestyle='dotted', linewidth=1)
plt.legend()
plt.show()
```

$\ddot {\theta} + \frac{g}{l} sen {\theta}$

---

$\theta = x_1$

$\dot{\theta} = x_2$

$\ddot{\theta} = \dot{x_2}$

---

$\ddot{x_2} = - \frac{g}{l} sen x_1$


```python
g = 9.81
l = 0.3
```

```python
def mydiff3(x,t,l,g):

  dx1dt = x[1]
  dx2dt = -(g/l) * np.sin(x[0])
  dxdt = [dx1dt,dx2dt] 
  return dxdt
```

```python
tst = -10
tend = 10
inc = 0.1
t = np.arange(tst,tend+inc,inc)
xinit = [np.deg2rad(45),0]
x = odeint(mydiff3,xinit,t,args=(l,g,))
```

```python
x1 = x[:,0]
x2 = x[:,1]
figure = plt.figure(figsize=(10,7))
plt.plot(t,x1,label=r'$\theta$')
plt.plot(t,x2,label=r'$\dot{\theta}$',color='purple')
plt.grid(color='r', linestyle='dotted', linewidth=1)
plt.legend()
plt.show()
```
